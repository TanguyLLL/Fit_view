```python
import json
import pandas as pd
import os
import sys
import matplotlib.pyplot as plt
import seaborn as sns
import ast
import numpy as np

sns.set(style="darkgrid")
%matplotlib inline
```

```python
# Usefull paths
fitbit_path = '/Users/tanguylienart/Documents/raspberry_projects/data/Tanguy/user-site-export/'
data_path= '/Users/tanguylienart/Documents/raspberry_projects/data/'  # Path used to save df
```

```python
heart_rate_chunk = pd.read_csv(data_path + 'heart_rate_uncorrected.csv', sep=';', chunksize = 25000, low_memory=False)
```

```python
i= 0
heart_rate_final = pd.DataFrame()
for chunk in heart_rate_chunk:
    i+=1
    if i% 10 ==0:
        print('{0} chunks processed'.format(i))

    chunk['value'] = chunk['value'].apply(lambda x : ast.literal_eval(x))
    chunk_value = chunk['value'].apply(pd.Series)

    chunk_date = chunk['dateTime']
    chunk_final = chunk_value.join(chunk_date)
    heart_rate_final = heart_rate_final.append(chunk_final)
```
